```python
import pandas_oracle.tools as pt # pandas의 oracle DB 연결하는 패키지
import os # 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
os.putenv('NLS_LANG', '.UTF8')
conn = pt.open_connection('../oracle\config.yml')
```

```python
#x :서비스 타입
#y :연령대별 판매건수

query1 = '''
SELECT *
FROM(SELECT 
       S_LOCATION,
       AVG(TS_MON_AMT) TS_MON_AMT,
       AVG(TS_TUE_AMT) TS_TUE_AMT,
       AVG(TS_WED_AMT) TS_WED_AMT,
       AVG(TS_THU_AMT) TS_THU_AMT,
       AVG(TS_FRI_AMT) TS_FRI_AMT,
       AVG(TS_SAT_AMT) TS_SAT_AMT,
       AVG(TS_SUN_AMT) TS_SUN_AMT,
       (AVG(TS_MON_AMT)+AVG(TS_TUE_AMT)+AVG(TS_WED_AMT)+AVG(TS_THU_AMT)+AVG(TS_FRI_AMT)+AVG(TS_SAT_AMT)+AVG(TS_SUN_AMT))/7 SUMAVG
       
FROM DAY_AMT_ANALYSIS T1, 
     ASTORE T2, 
     ASERVICE T3 
WHERE (T1.S_CODE = T2.S_CODE)
      and (T1.SE_CODE = T3.SE_CODE)
      and T1.TS_QUARTER = 1
      
Group By S_LOCATION
Order By SUMAVG DESC) a1
WHERE ROWNUM BETWEEN 1 AND 5
'''
```
