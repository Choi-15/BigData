```python
import pandas_oracle.tools as pt # pandas의 oracle DB 연결하는 패키지
import os # 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
# putenv(key,value) : NLS_LANG(SQL에서의 환경변수)의 언어설정을 UTF-8로 설정
os.putenv('NLS_LANG', '.UTF8') # 해당 호출의 자동변환을 위한 메소드
# conn 객체에 연결 설정파일 저장
conn = pt.open_connection('../pandas/config.yml')

# query 객체에 사용하고자 하는 테이블(뷰) 저장
query = '''
SELECT *

FROM (
    SELECT S_LOCATION,
       round(avg(TS_10_NUM) + avg(TS_20_NUM) + avg(TS_30_NUM) + avg(TS_40_NUM) + avg(TS_50_NUM) + avg(TS_60UP_NUM)) SUM_NUM , 
       round(avg(TS_10_NUM)) AGE_10_AVG,
       round(avg(TS_20_NUM)) AGE_20_AVG,
       round(avg(TS_30_NUM)) AGE_30_AVG,
       round(avg(TS_40_NUM)) AGE_40_AVG,
       round(avg(TS_50_NUM)) AGE_50_AVG,
       round(avg(TS_60UP_NUM)) AGE_60UP_AVG
       
    FROM ASTORE a, TOTALSELL t
    WHERE a.S_CODE = t.S_CODE
    GROUP BY S_LOCATION
    ORDER BY SUM_NUM DESC)
'''

# pt.query_to_df(query, conn_db, arraysize) : DataFrame으로 형변환하는 메소드
# query:사용하고자하는 쿼리
# conn_db: cx_oracle 또는 open_connection의 연결 객체
# arraysize: 가져올 쿼리의 arraysize 지정
df = pt.query_to_df(query, conn, 1000000)
# df_ageAmount의 스키마 구조 파악
print(df.info())
print('-'*80)
# df_ageAmount 데이터프레임 보기
print(df)
```

```python
df_ageNum = df.head()
df_ageNum
```

```python
# x : 상권 위치
# y :연령대별 판매건수

plt.bar(df_ageNum.S_LOCATION, df_ageNum.AGE_10_AVG, label='10대')
plt.bar(df_ageNum.S_LOCATION, df_ageNum.AGE_20_AVG, label='20대', bottom=df_ageNum.AGE_10_AVG)
plt.bar(df_ageNum.S_LOCATION, df_ageNum.AGE_30_AVG, label='30대', bottom=df_ageNum.AGE_10_AVG+df_ageNum.AGE_20_AVG)
plt.bar(df_ageNum.S_LOCATION, df_ageNum.AGE_40_AVG, label='40대', bottom=df_ageNum.AGE_10_AVG+df_ageNum.AGE_20_AVG+df_ageNum.AGE_30_AVG)
plt.bar(df_ageNum.S_LOCATION, df_ageNum.AGE_50_AVG, label='50대', bottom=df_ageNum.AGE_10_AVG+df_ageNum.AGE_20_AVG+df_ageNum.AGE_30_AVG+df_ageNum.AGE_40_AVG)
plt.bar(df_ageNum.S_LOCATION, df_ageNum.AGE_60UP_AVG, label='60대~', bottom=df_ageNum.AGE_10_AVG+df_ageNum.AGE_20_AVG+df_ageNum.AGE_30_AVG+df_ageNum.AGE_40_AVG+df_ageNum.AGE_50_AVG)
plt.legend(loc='upper right')

plt.title("연령대별 지역의 당월 판매 건수")
plt.xlabel("지역")
plt.ylabel("판매 건수")
plt.show()
```
